```python
import string

def remove_punctuation(input_string):
    # Make a translator object to remove punctuation
    translator = str.maketrans('', '', string.punctuation)

    # Use the translator
    no_punct = input_string.translate(translator)
    return no_punct
```

```python
def preprocessing_lyric(s):
    return remove_punctuation(s)
```

```python
def preprocessing_reference(ls, special_dict):
    new_l = []
    for l in ls:
        if l in special_dict:
            l = special_dict[l]
        l = l.replace(". ", ".")
        l = l.replace(". ", ".")
        l = l.replace(" - ", "-")
        l = l.replace("- ", "-")
        l = l.replace(" -", "-")
        # print(l)
        pattern2 = r"(.+[a-zA-Z]+)[\s:\.]?(\d+)[\s:\.](.+)"
        m = re.search(pattern2, l)
        # print(m.groups())
        # print(m.groups()[0], m.groups()[1], m.groups()[2])
        try:
            new_l.append(m.groups()[0] + " " +  m.groups()[1] + '.' + m.groups()[2])
        except:
            print(ls)
    return new_l
```

```python
import re
def fill_list(l, bible):
    verse = ""
    temp = l.split(' ')
    if len(temp) == 3:
        b = temp[0] + temp[1]
        c_v = temp[2]
    elif len(temp) == 2:
        b = temp[0]
        c_v = temp[1]
    else:
        print("issue", temp)
    temp = c_v.split(".")

    try:
        c = int(temp[0])
        v = temp[1]
    except:
        print(l)
    if '–' in v:
        v = v.replace('–', '-')
    if "-" in v:
        temp = v.split('-')
        start_v = temp[0]
        end_v = temp[1]
        for v_ in range(clean_verse_number(start_v), clean_verse_number(end_v)+1):
            verse += bible.get_verse(check_book_name(b), c, v_)
    else:
        verse += bible.get_verse(check_book_name(b), c, clean_verse_number(v))
    return verse
```

```python
def clean_verse_number(verse_string):
    result = re.sub(r"[^0-9]", "", verse_string)
    return int(result)
```

```python
class Bible():
    def __init__(self):
        self.df = pd.read_csv("./drive/MyDrive/bible_data_set.csv")
        self.verse_group = self.df.groupby(['book','chapter','verse'])
        self.chapter_group = self.df.groupby(['book','chapter'])
        self.df.text = self.df.text.apply(preprocessing_lyric)
        
    def get_verse(self, book, chapter, verse):
        try:
            return self.verse_group.get_group((book, chapter, verse)).text.iloc[0].strip(' \n')
        except:
            return ''
    
    def get_verse_num(self, book, chapter):
        return len(self.chapter_group.get_group((book, chapter)))
```

```python
def assign_label(qid, docno, inverted_index, inverted_index_theme_filtered, theme_l_set):
    qid = int(qid)
    docno = int(docno)
    if docno in inverted_index[qid]:
        return 3
    elif (len(inverted_index_theme_filtered[qid] & theme_l_set[docno]) >= 2):
        return 2
    elif (len(inverted_index_theme_filtered[qid] & theme_l_set[docno]) == 1):
        return 1
    else:
        return 0
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
from sklearn.metrics.pairwise import cosine_similarity

#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')


def get_embedding_similarity(sentences, tokenizer, model):
    # Tokenize sentences
    encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input)

    # Perform pooling
    sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

    # Normalize embeddings
    sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
    # print(sentence_embeddings)
    # print(type(sentence_embeddings))
    # print(type(sentence_embeddings[0]))
    # compute the cosine similarity
    cos = torch.nn.CosineSimilarity(dim=0)
    similarity = cos(sentence_embeddings[0], sentence_embeddings[1])

    return similarity
```
