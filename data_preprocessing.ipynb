```python
songs = songs.dropna(subset=['lyric'])
songs = songs[songs['themes'] != '[]']
songs = songs[songs['reference'] != '[]']
```

```python
songs['title'] = songs['title'].apply(lambda x:x.strip())
songs['lyric'] = songs['lyric'].apply(lambda x:x.strip())
```

```python
songs['themes'] = songs['themes'].apply(lambda x:x.replace('\'', '').replace('\"', '').strip('][').split(', '))
```

```python
songs['reference'] = songs['reference'].apply(lambda x:x.replace('\'', '').replace('\"', '').strip('][').split(', '))
```

```python
songs = songs[songs['themes'].map(lambda d: len(d)) > 0]
songs = songs[songs['reference'].map(lambda d: len(d)) > 0]
```

```python
songs = songs.drop_duplicates(['lyric'])
```
